# Prática 01: Implementação de RPC's!

```python
class Somador:
   def sum(self, vetor):
        soma = 0
        for i in vetor:
            soma += i
        
        return soma
```

```python
class Server:
    def __init__(self):
        self.somador = Somador()
        
    def query_sum(self, vetor):
        resposta = self.somador.sum(vetor)
        return resposta
        
```

```python
class Client:
    def __init__(self, servidor_conectado):
        self.servidor = servidor_conectado
    
    def requisicao_sum(self, vetor):
            return self.servidor.query_sum(vetor)
```

```python
servidor = Server()
cliente = Client(servidor)

vetor = [1,2,3,4,5]

cliente.requisicao_sum(vetor)
```
